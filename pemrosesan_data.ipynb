## Translated Text

```python
data = pd.read_csv("data_test_40000.csv", sep=',')

data['clean_text'] = data['data_text'].str.replace('([^0-9A-Za-z \t])|(\w+:\/\/\S+)', ' ')
data['clean_text'] = data['clean_text'].str.lower()

data.head(20)
```

```python
import time
from googletrans import Translator  

# Inisiasi penerjemah
translator = Translator()

# Fungsi untuk menerjemahkan teks individual
def translate_text(text, translator):
    try:
        # Hanya terjemahkan jika teks bukan string kosong
        if isinstance(text, str) and text.strip() != "":
            translated = translator.translate(text, dest='en').text
            return translated
        return text
    except Exception as e:
        print(f"Error saat menerjemahkan: {e}")
        return text  # Kembalikan teks asli jika gagal diterjemahkan

# Fungsi untuk menerjemahkan teks dalam batch
def translate_text_batch(data_column, translator, batch_size=500, sleep_time=2):
    translated_texts = []
    for i in range(0, len(data_column), batch_size):
        batch = data_column[i:i + batch_size]
        for text in batch:
            translated_texts.append(translate_text(text, translator))
        print(f"Translated batch {i // batch_size + 1}")
        time.sleep(sleep_time)  # Tambahkan jeda 2 detik setelah setiap batch untuk menghindari rate limit
    return translated_texts

# Jalankan penerjemahan dalam batch kecil
try:
    data['translated_text'] = translate_text_batch(data['clean_text'], translator, batch_size=400, sleep_time=2)
    print(data.head(20))
except KeyboardInterrupt:
    print("Process was interrupted by the user.")

# Lihat hasil 20 baris pertama
print(data.head(20))
data.to_csv('translated_data_40000.csv', encoding='utf8', index=False)
```

## Labeling Data

```python
# Importing required libraries
import pandas as pd
from textblob import TextBlob
import matplotlib.pyplot as plt
from nltk.corpus import stopwords
import nltk

# Download stopwords if not already downloaded
nltk.download('stopwords')

# Load data from 'translated_data_40000.csv'
data = pd.read_csv('translated_data_40000.csv')

# Combine all text into a single string
all_text = ' '.join(data['clean_text'].astype(str))

# Use English stopwords
stop_words = set(stopwords.words('english'))

# Remove stop words from the combined text
all_text_no_stopwords = ' '.join([word for word in all_text.split() if word.lower() not in stop_words])

# Function to calculate subjectivity
def subjektivitas(tr_text):
    return TextBlob(tr_text).sentiment.subjectivity

# Function to calculate polarity (sentiment)
def polaritas(tr_text):
    return TextBlob(tr_text).sentiment.polarity

# Function to assign sentiment based on polarity
def hasil_sentimen(nilai):
    if nilai < 0:
        return 'negatif'
    else:
        return 'positif'

# Function to assign topic label based on keywords
def label_comment(text):
    keywords = {
        'Respon Emosional Penggemar': [
            'love', 'support', 'fan', 'admire', 'idol', 'respect', 'goat', 'legend', 'hero', 'admiration', 
            'passion', 'loyalty', 'devotion', 'appreciate', 'care', 'trust', 'favorite', 'inspire', 'motivate', 
            'dedicate', 'affection', 'heart', 'dedication', 'adoration', 'devoted', 'grateful', 'pride', 
            'emotional', 'tears', 'cheer', 'thankful', 'bless', 'joy', 'honor', 'congratul', 'bro', 'best', 
            'hello', 'wow', 'great', 'king', 'greet', 'happi', 'beat', 'amazing', 'famili', 'deserv', 'children', 
            'grateful', 'cry', 'crazi', 'thank'
        ],
        'Interaksi dengan Platform YouTube': [
            'subscribe', 'youtube', 'creator', 'channel', 'views', 'like', 'watch', 'comment', 'content', 
            'share', 'viral', 'engage', 'engagement', 'upload', 'post', 'trend', 'stream', 'video', 'clips', 
            'reaction', 'algorithm', 'platform', 'follower', 'subscriber', 'hit', 'popular', 'growth', 'reach', 
            'audience', 'notification', 'monetize', 'influencer', 'analytics', 'subscribe button', 'button', 
            'sub', 'get', 'record', 'view', 'goal', 'share', 'notif', 'stream', 'clip', 'live', 'pl', 'hit', 'like'
        ],
        'Pengaruh Global Ronaldo': [
            'world', 'global', 'india', 'bangladesh', 'international', 'indian', 'fans', 'community', 'culture', 
            'nation', 'country', 'continent', 'asia', 'europe', 'africa', 'america', 'supporters', 'foreign', 
            'worldwide', 'reach', 'global star', 'global icon', 'cross-border', 'multinational', 'impact', 
            'popularity', 'famous', 'celebrity', 'universal', 'cross-cultural', 'expansion', 'international fame', 
            'diverse', 'ethnic', 'nationwide', 'global market', 'brand', 'ambassador', 'representative', 
            'international appeal', 'cosmopolitan', 'indonesia', 'mr', 'beast', 'messi', 'ronaldo', 'christian', 
            'supporters', 'internationally', 'madrid', 'god', 'real', 'legend', 'indonesia'
        ]
    }
    
    for label, words in keywords.items():
        for word in words:
            if word.lower() in text.lower():
                return label
    return 'Unlabeled'

# Adding columns for subjectivity, polarity, sentiment, and topic label
data['subjektivitas'] = data['translated_text'].apply(subjektivitas)
data['polaritas'] = data['translated_text'].apply(polaritas)
data['sentimen'] = data['polaritas'].apply(hasil_sentimen)
data['label'] = data['translated_text'].apply(label_comment)

# Grouping data by label and sentiment for visualization
grouped = data.groupby(['label', 'sentimen']).size().unstack(fill_value=0)

# Display sentiment counts for each topic
print(grouped)

# Plotting the data as a bar chart
grouped.plot(kind='bar', stacked=True)
plt.title('Jumlah Sentimen Positif dan Negatif untuk Setiap Topik')
plt.xlabel('Topik')
plt.ylabel('Jumlah')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Save the updated data to a new CSV file
data.to_csv('done_data_40000.csv', encoding='utf8', index=False)
```
